```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target
def plot_iris(X, y):
    fig = plt.figure(figsize=(10,4))
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Iris setosa")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Iris versicolor")
    plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="Iris virginica")
    return fig
```

```python
for depth in range(1,6):
    plot_iris(X,y)
    axes = plt.axis()
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    tree = DecisionTree.DecisionTree(max_depth=depth,random_state=1122)
    tree.fit(X,y)
    y_pred = np.array(tree.predict(X_new)).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.5, cmap=custom_cmap)
    plt.title('depth = {}'.format(depth))
    plt.show()
```

```python
np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
y = y + np.random.randn(m, 1) / 10
```

```python
def plot_regression(X,y):
    plt.xlabel("$x_1$", fontsize=18)
    plt.plot(X, y, "b.")
#plot_regression(X,y)
```

```python
from Trees import DecisionTree
from importlib import reload as rl
rl(DecisionTree)
from Trees import DecisionTree
tree  = DecisionTree.RegressionTree(max_depth=1)
tree.fit(X,y)
```

```python
for depth in range(1,15):
    plot_regression(X,y)
    axes = plt.axis()
    x1 = np.linspace(axes[0], axes[1], 500).reshape(-1, 1)
    tree = DecisionTree.RegressionTree(max_depth=depth)
    tree.fit(X,y)
    plt.title('depth= {}'.format(depth))
    y_pred = tree.predict(x1)
    plt.plot(x1, y_pred, "r-", linewidth=1, label=r"$\hat{y}$")
    plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
df = pd.DataFrame(housing['data'], columns = housing['feature_names'])
df['target'] = housing['target']
df.head()
```

```python
df.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4,
             s=df["Population"]/100, label="Population", figsize=(10,7),
             c="target", cmap=plt.get_cmap("jet"), colorbar=True,
             sharex=False)
plt.legend()
```

```python
for depth in range(1,20,5):
    tree = DecisionTree.RegressionTree(max_depth=depth, splitter='quantile')
    tree.fit(df.iloc[:,:-1].to_numpy(), df.iloc[:,-1].to_numpy())
    y_pred = tree.predict(df.iloc[:,:-1].to_numpy())
    fig, axes = plt.subplots(nrows = 1, ncols=3,figsize = (20,4))
    df.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4,
                 #s=df["Population"]/100, label="Population", 
                    #figsize=(8,4),
                ax = axes[0],
                 c= 'target', cmap=plt.get_cmap("jet"), colorbar=True,
                 sharex=False,title = 'True')

    df.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4,
                 #s=df["Population"]/100, label="Population", 
                    #figsize=(8,4),
            title = 'depth= {}'.format(depth),
                ax = axes[1],
                 c= y_pred, cmap=plt.get_cmap("jet"), colorbar=True,
                 sharex=False)
    df.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4,
             #s=df["Population"]/100, label="Population", 
                #figsize=(8,4),
        title = 'Error'.format(depth),
            ax = axes[2],
             c= abs(y_pred-df['target']), cmap=plt.get_cmap("PuRd"), colorbar=True,
             sharex=False)
    #plt.legend()
    plt.show()
```
